```python
# 5,7,20,23
cols_to_scale = ['V5', 'V7', 'V20', 'V23']

scaler = RobustScaler()
df_scaled_values = scaler.fit_transform(df[cols_to_scale])

df_scaled_values

df_scaled = df.copy()
df_scaled[cols_to_scale] = df_scaled_values

df_scaled.head()
```

```python
sns.boxplot(data=df_scaled.loc[:, 'V21':'V28'])
```

```python
X = df_scaled.loc[:, 'V1':'V28']
X.head()
```

```python
y = df_scaled['Class']
y.head()
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(X, y, test_size= 0.33, random_state = 42)

model = XGBClassifier(scale_pos_weight=100, use_label_encoder=False, eval_metric='aucpr')
model.fit(xtrain, ytrain)
```

```python
ypred = model.predict(xtest)

print("Accuracy: ", accuracy_score(ytest, ypred)*100)
```
